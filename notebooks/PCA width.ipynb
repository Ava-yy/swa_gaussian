```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = np.load('../experiments/width/resnet110_pca_width.npz')
```

```python
N = f['N']
dim = f['dim']
pc_ids = f['pc_idx']
evr = f['explained_variance_ratio']
ts = np.linspace(-150, 150, 31) # f['ts']
test_err = f['test_err']
test_loss = f['test_loss']
train_err = f['train_err']
train_loss = f['train_loss']
```

```python
plt.figure(figsize=(12, 7))
for i in range(len(pc_ids)):
    pc_id = pc_ids[i]
    plt.plot(ts, test_err[i, :], linewidth=2, label='PC #%3d [%6.2f %%]' % (pc_id + 1, evr[pc_id] * 100))
plt.xlabel('Distance', fontsize=24)
plt.xticks(fontsize=20)
plt.ylabel('Test Error (%)', fontsize=24)
plt.yticks(fontsize=20)
plt.legend(fontsize=18, bbox_to_anchor=(1.02, 0.99))
plt.title('PCA ResNet110 CIFAR-100\n Dimension: %d\n SGD samples: %d' % (dim, N), fontsize=24)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
for i in range(len(pc_ids)):
    pc_id = pc_ids[i]
    plt.plot(ts, test_loss[i, :], linewidth=2, label='PC #%3d [Var=%.2f %%]' % (pc_id + 1, evr[pc_id] * 100))
plt.xlabel('Distance', fontsize=24)
plt.xticks(fontsize=20)
plt.ylabel('Test loss', fontsize=24)
plt.yticks(fontsize=20)
plt.legend(fontsize=18, bbox_to_anchor=(1.02, 0.99))
plt.title('PCA ResNet110 CIFAR-100\n Dimension: %d\n SGD samples: %d' % (dim, N), fontsize=24)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
for i in range(len(pc_ids)):
    pc_id = pc_ids[i]
    plt.plot(ts, train_err[i, :], linewidth=2, label='PC #%3d [%6.2f %%]' % (pc_id + 1, evr[pc_id] * 100))
plt.xlabel('Distance', fontsize=24)
plt.xticks(fontsize=20)
plt.ylabel('Train Error (%)', fontsize=24)
plt.yticks(fontsize=20)
plt.legend(fontsize=18, bbox_to_anchor=(1.02, 0.99))
plt.title('PCA ResNet110 CIFAR-100\n Dimension: %d\n SGD samples: %d' % (dim, N), fontsize=24)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
for i in range(len(pc_ids)):
    pc_id = pc_ids[i]
    plt.plot(ts, train_loss[i, :], linewidth=2, label='PC #%3d [Var=%.2f %%]' % (pc_id + 1, evr[pc_id] * 100))
plt.xlabel('Distance', fontsize=24)
plt.xticks(fontsize=20)
plt.ylabel('Train loss', fontsize=24)
plt.yticks(fontsize=20)
plt.legend(fontsize=18, bbox_to_anchor=(1.02, 0.99))
plt.title('PCA ResNet110 CIFAR-100\n Dimension: %d\n SGD samples: %d' % (dim, N), fontsize=24)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
w = 0.7
plt.bar(1 + np.arange(60) - 0.5 * w, evr[:60] * 100, width=w)
plt.xlabel('# of PC', fontsize=24)
plt.xticks(fontsize=20)
plt.ylabel('Explained variance ratio (%)', fontsize=24)
plt.yticks(fontsize=20)
plt.show()
```
