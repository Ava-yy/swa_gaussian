```python
class LogNormalize(colors.Normalize):    
    def __init__(self, vmin=None, vmax=None, clip=None, log_alpha=None):        
        self.log_alpha = log_alpha
        colors.Normalize.__init__(self, vmin, vmax, clip)
    def __call__(self, value, clip=None):    
        log_v = np.ma.log(value - self.vmin)
        log_v = np.ma.maximum(log_v, self.log_alpha)
        return 0.9 * (log_v - self.log_alpha) / (np.log(self.vmax - self.vmin) - self.log_alpha)

def plane(grid, M, vmax=None, log_alpha=-5, N=7, cmap='jet_r'):
    cmap = plt.get_cmap(cmap)
    if vmax is None:
        C = M.copy()
    else:
        C = np.minimum(M, vmax)
    log_gamma = (np.log(C.max() - C.min()) - log_alpha) / N
    levels = C.min() + np.exp(log_alpha + log_gamma * np.arange(N + 1))
    levels[0] = C.min() + 1e-8
    levels[-1] = C.max() + 1e-8
    levels = np.concatenate((levels, [1e10]))
    norm = LogNormalize(C.min(), C.max(), log_alpha=log_alpha)
    plt.contour(grid[:, :, 0], grid[:, :, 1], M, cmap=cmap, norm=norm, linewidths=2.5, zorder=3, levels=levels)          
    cf = plt.contourf(grid[:, :, 0], grid[:, :, 1], M, cmap=cmap, norm=norm, levels=levels, zorder=0, alpha=0.55)      
    cb = plt.colorbar(format='%.2g') 
    labels = list(cb.ax.get_yticklabels())    
    labels[-1].set_text(r'$>\,$' + labels[-2].get_text())
    cb.ax.set_yticklabels(labels)
    return cb 
```

```python
f = np.load('../experiments/width/resnet110_pca_2d.npz')
xs, ys = f['xs'], f['ys']
grid = np.stack(np.meshgrid(xs, ys), axis=-1)
train_loss = f['train_loss']
ev = f['explained_variance']
pc1_id = f['pc1_id']
pc2_id = f['pc2_id']
trajectory_projection = f['trajectory_projection']
plt.figure(figsize=(12, 7))
cb = plane(grid, train_loss.T, vmax=5.0, log_alpha=-4)
cb.ax.tick_params(labelsize=18)

t = np.linspace(0, 2.0 * np.pi, 100)
ex = 3.0 * np.sqrt(ev[pc1_id]) * np.cos(t)
ey = 3.0 * np.sqrt(ev[pc2_id]) * np.sin(t)
plt.plot([0], [0], 'k*', markersize=15, label='SWA')
plt.plot(ex, ey, 'k--', linewidth=2, dashes=(6, 8), label='Empirical $3\sigma$')
plt.plot(trajectory_projection[0, :], trajectory_projection[1, :], 'c--', label='Proj(Trajectory)')
plt.legend(fontsize=18, bbox_to_anchor=(1.64, 1.05))
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.title('PCA ResNet110 CIFAR-100', fontsize=24, y=1.02)
plt.grid()
plt.show()
```

```python
import torch
import sys
sys.path.append('..')
import swag.models
from swag.posteriors.swag import SWAG
import sklearn.decomposition

model = swag.models.VGG16.base(num_classes=100).cuda()
swag_diag = SWAG(swag.models.VGG16.base, num_classes=100, no_cov_mat=True).cuda()
swag_cov = SWAG(swag.models.VGG16.base, num_classes=100, no_cov_mat=False, max_num_models=20).cuda()

W = []
for i in range(161, 301):
    path = '../experiments/train/ckpts/c100/vgg16/run1/swag/checkpoint-%d.pt' % i
    checkpoint = torch.load(path)
    model.load_state_dict(checkpoint['state_dict'])
    swag_diag.collect_model(model)
    swag_cov.collect_model(model)
    
    W.append(np.concatenate([p.detach().cpu().numpy().ravel() for p in model.parameters()]))
W = np.array(W)
```

```python
mean, _, cov_mat_list = swag_cov.export_numpy_params(export_cov_mat=True)
cov_mat = np.hstack([m.reshape(20, -1) for m in cov_mat_list])

np.random.seed(1)
tsvd = sklearn.decomposition.TruncatedSVD(n_components=20, n_iter=7)
tsvd.fit(cov_mat)
```

```python
u = tsvd.components_[0, :].copy()
u /= np.linalg.norm(u)
v = tsvd.components_[1, :].copy()
v /= np.linalg.norm(v)

mean_diag, variances_diag = swag_diag.export_numpy_params()

ex_diag = np.sum(u * variances_diag * u)
ey_diag = np.sum(v * variances_diag * v)
exy_diag = np.sum(u * variances_diag * v)
print(np.sqrt(ex_diag), np.sqrt(ey_diag), exy_diag)
```
