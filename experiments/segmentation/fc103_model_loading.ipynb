```python
model_weights = np.load('/home/wesley/Documents/Code/FC-DenseNet/weights/FC-DenseNet103_weights.npz')
with model_weights as f:
    saved_param_values = [f['arr_%d' % i] for i in range(len(f.files))]
```

```python
#only gradient based parameters
model_parameters = list(model.named_parameters())
```

```python
len(model_parameters)
```

```python
len(saved_param_values)
```

```python
zero_layers = []
one_layers = []
for p in saved_param_values:
    zero_layers.append(p.sum()==0)
    one_layers.append(p.mean()==1)
bad_layers = [z or o for z, o in zip(zero_layers, one_layers)]
```

```python
no_bad_params = []
for param, isbad in zip(saved_param_values, bad_layers):
    if not isbad:
        no_bad_params.append( param )
```

```python
len(no_bad_params)
```

```python
theano_final_params = 0
for param in no_bad_params:
    theano_final_params += param.size
print(theano_final_params)

pytorch_params = 0
for _, param in model_parameters:
    pytorch_params += param.numel()
print(pytorch_params)
```

```python
params_eq = []
for i, (named_parameter, weight_to_load) in enumerate(zip(model_parameters, no_bad_params)):
    
    is_eq = list(named_parameter[1].size()) == list(weight_to_load.shape)
    params_eq.append(is_eq)
    if not is_eq:
        print(i)
        print(named_parameter[1].size(), weight_to_load.shape)
        print(named_parameter[0])
    
```
