```python
def softmax(x, axis=1):
    return np.exp(x) / np.sum(np.exp(x),axis=axis, keepdims=True)
```

```python
Sky = [128,128,128]
Building = [128,0,0]
Pole = [192,192,128]
Road = [128,64,128]
Pavement = [60,40,222]
Tree = [128,128,0]
SignSymbol = [192,128,128]
Fence = [64,64,128]
Car = [64,0,128]
Pedestrian = [64,64,0]
Bicyclist = [0,128,192]
Unlabelled = [0,0,0]

label_colours = np.array([Sky, Building, Pole, Road, Pavement,
            Tree, SignSymbol, Fence, Car, Pedestrian, Bicyclist, Unlabelled])
```

```python
def view_annotated(temp, alpha=None, plot=True):
    #emp = tensor.numpy()
    r = temp.copy()
    g = temp.copy()
    b = temp.copy()
    for l in range(0,11):
        r[temp==l]=label_colours[l,0]
        g[temp==l]=label_colours[l,1]
        b[temp==l]=label_colours[l,2]

    rgb = np.zeros((temp.shape[0], temp.shape[1], 3))
    rgb[:,:,0] = (r/255.0)#[:,:,0]
    rgb[:,:,1] = (g/255.0)#[:,:,1]
    rgb[:,:,2] = (b/255.0)#[:,:,2]
    print(rgb.shape)
    
    if alpha is not None:
        rgb = np.append(rgb, np.expand_dims(alpha, 2), 2)
        
    if plot:
        plt.imshow(rgb)
        plt.show()
    else:
        return rgb
```

## Calibration

```python
def compute_accuracy_by_score(score, preds, labels, bins=None, background_class=11):
    labels = labels.reshape(-1)
    
    background_labels = labels == background_class
    
    labels = labels[~background_labels]
    
    score = score.reshape(-1)
    score = score[~background_labels]
    preds = preds.reshape(-1)
    preds = preds[~background_labels]
    
    if bins is None:
        bins = np.percentile(score, np.arange(100))
        
    digitized = np.digitize(score, bins)
    accuracy = [np.sum(preds[digitized==i]==labels[digitized==i])/np.sum(digitized==i) for i in range(1, len(bins))]
    return accuracy, bins
```

## Mutual Information

```python

for i in range(14):
    print(i)
    f_loc = 'swag_preds_testpred_'+str(i)+'.npz'
    fdict = np.load(f_loc)
    
    if i == 0:
        neg_entropies = np.sum(np.log(fdict['predictions'] + 1e-7) * fdict['predictions'], axis=1)
        predictions = fdict['predictions']
        
    else:
        neg_entropies += np.sum(np.log(fdict['predictions'] + 1e-7) * fdict['predictions'], axis=1)
        predictions += fdict['predictions']
```

```python
predictions /= 14
neg_entropies /= 14

entropy = -np.sum(np.log(predictions + 1e-7) * predictions, axis = 1)

mutual_info = entropy + neg_entropies
```

```python
for i in range(14):
    print(i)
    f_loc = 'old_testpreds/dropout_preds_testpred_'+str(i)+'.npz'
    fdict = np.load(f_loc)
    
    if i == 0:
        dneg_entropies = np.sum(np.log(fdict['predictions'] + 1e-7) * fdict['predictions'], axis=1)
        dpredictions = fdict['predictions']
        
    else:
        dneg_entropies += np.sum(np.log(fdict['predictions'] + 1e-7) * fdict['predictions'], axis=1)
        dpredictions += fdict['predictions']
```

```python
dpredictions /= 14
dneg_entropies /= 14

dentropy = -np.sum(np.log(dpredictions + 1e-7) * dpredictions, axis = 1)

dmutual_info = dentropy + dneg_entropies
```
